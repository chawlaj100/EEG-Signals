```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df_temp = df.copy()
```

```python
df_af3 = df_temp[df_temp['Channel'] == 'AF3']
df_af4 = df_temp[df_temp['Channel'] == 'AF4']
df_t7 = df_temp[df_temp['Channel'] == 'T7']
df_t8 = df_temp[df_temp['Channel'] == 'T8']
df_pz = df_temp[df_temp['Channel'] == 'PZ']
```

```python
df_pz.head()
```

```python
df_pz[df_pz['Digit'] == 1]
```

```python
fig = plt.figure(figsize=(20,15))
plt.ylim(4150, 4350)
plt.scatter(df_pz['Event'], df_pz['Avg_value'])
plt.xlabel("Digits")
plt.ylabel("Value")
plt.show
```

```python
fig = plt.figure(figsize=(20,15))
plt.ylim(4200, 4400)
plt.scatter(df_af3['Digit'], df_af3['Avg_value'])
#plt.scatter(df_af3['Digit'], df_af3['Avg_value'], color='red')
plt.xlabel("Digits")
plt.ylabel("Value")
plt.show
```

```python
sns.set(rc={'figure.figsize':(20,15)})
ax = sns.violinplot(x = df_af3['Digit'], y = df_af3['Avg_value'], inner='quartile', grid=True)
ax.set_ylim(4250,4350)
```

```python
sns.set(rc={'figure.figsize':(20,15)})
ax = sns.violinplot(x = df_pz['Digit'], y = df_pz['Avg_value'], grid=True)
ax.set_ylim(4150,4350)
```

```python
df_slice = df_temp[['Digit', 'Avg_value']]
```

```python
arr = np.array(df_slice, dtype="int")
arr[0:5,1]
```

```python
values = []
digit = []
for i in range(0, 65250, 5):
    values.append(sum(arr[i:i+5,1])/5)
    digit.append(arr[i,0])
```

```python
values = np.array(values)
digit = np.array(digit)
values = values.reshape(-1,1)
digit = digit.reshape(-1,1)
total = np.append(digit, values, axis=1)
total.shape
```

```python
total
```

```python
sns.set(rc={'figure.figsize':(20,15)})
ax = sns.violinplot(x = total[:,0], y = total[:,1], inner='quartile')
ax.set_ylim(4150,4350)
```
