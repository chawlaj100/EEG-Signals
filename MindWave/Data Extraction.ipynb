```python
import os
import numpy as np
import pandas as pd
```

```python
temp = []
with open("MW.txt") as f:
    for row in f:
        arr = row.split("\t")[4:]
        t = arr[-1].split(",")
        t[-1] = t[-1][:-1]
        arr[2:] = t
        temp.append(arr)
```

```python
len(temp)
```

```python
temp_int = []
for t in temp:
    arr = list(map(int, t))
    temp_int.append(arr)
```

```python
temp_int[1]
```

```python
df = pd.DataFrame(temp_int)
df.to_csv("train.csv", index=False)
```
