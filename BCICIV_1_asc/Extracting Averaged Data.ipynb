```python
import os
import numpy as np
import pandas as pd
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_mrk.txt"
seq = ['a','b','c','d','e','f','g']
total = []
for i in seq:
    arr = []
    with open(file+i+end) as f:
        for row in f:
            time = row.split("\t")[0]
            arr.append(float(time))
    arr = list(map(int, arr))
    total.append(arr)
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_cnt.txt"
seq = ['a','b','c','d','e','f','g']
j = 0
result_X = []
for i in seq:
    rows = []
    with open(file+i+end) as f:
        indexes = total[j]
        for row in f:
            rows.append(row[:-1])
        for end_index in indexes:
            st_index = end_index - 400
            sum = np.zeros((59,))
            for i in range(st_index, end_index+1):
                r = rows[i]
                res = r.split('\t')
                res = list(map(int, res))
                res = np.array(res)
                sum = sum + res
            sum = sum//400
            arr = list(sum)
            result_X.append(arr)
    print("Done")
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_mrk.txt"
seq = ['a','b','c','d','e','f','g']
Y = []
for i in seq:
    with open(file+i+end) as f:
        for row in f:
            time = row.split("\t")[1]
            Y.append(float(time))        
Y = list(map(int, Y))
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_nfo.txt"
seq = ['a','b','c','d','e','f','g']
l = []
for i in seq:
    with open(file+i+end) as f:
        for row in f:
            if row.startswith("classes"):
                l.append(row[:-1].split(":")[1][1:])
```

```python
classes = []
for i in l:
    classes.append(i.split(","))
for c in classes:
    c[1] = c[1][1:]
classes
```

```python
Y_n = []
for i in range(1400):
    if i % 200 == 0:
        j = i//200
    in1, in2 = classes[j]
    if Y[i] == -1:
        Y_n.append(in1)
    else:
        Y_n.append(in2)
```

```python
Y_n = np.array(Y_n)
Y_n = Y_n.reshape(1400,1)
Y_n.shape
```

```python
train_data = np.append(X_n, Y_n, axis = 1)
```
