```python
import os
import numpy as np
import pandas as pd
```

```python
files = os.listdir("Data")
files[:5]
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_mrk.txt"
seq = ['a','b','c','d','e','f','g']
total = []
for i in seq:
    arr = []
    with open(file+i+end) as f:
        for row in f:
            time = row.split("\t")[0]
            arr.append(float(time))
    arr = list(map(int, arr))
    total.append(arr)
```

```python
len(total)
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_cnt.txt"
seq = ['a','b','c','d','e','f','g']
X = []
j = 0
for i in seq:
    with open(file+i+end) as f:
        indices = total[j]
        val = []
        for row in f:
            val.append(row[:-1])
        for index in indices:
            X.append(val[index].split("\t"))
X
```

```python
X_n = []
for l in X:
    a = []
    for b in l:
        a.append(int(b))
    X_n.append(a)
print(len(X_n))
X_np = np.array(X_n, dtype="int")
X_np[0]
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_mrk.txt"
seq = ['a','b','c','d','e','f','g']
Y = []
for i in seq:
    with open(file+i+end) as f:
        for row in f:
            time = row.split("\t")[1]
            Y.append(float(time))        
Y = list(map(int, Y))
```

```python
len(Y)
```

```python
file = "Data/BCICIV_calib_ds1"
end = "_nfo.txt"
seq = ['a','b','c','d','e','f','g']
l = []
for i in seq:
    with open(file+i+end) as f:
        for row in f:
            if row.startswith("classes"):
                l.append(row[:-1].split(":")[1][1:])
```

```python
classes = []
for i in l:
    classes.append(i.split(","))
for c in classes:
    c[1] = c[1][1:]
classes
```

```python
Y_n = []
for i in range(1400):
    if i % 200 == 0:
        j = i//200
    in1, in2 = classes[j]
    if Y[i] == -1:
        Y_n.append(in1)
    else:
        Y_n.append(in2)
```

```python
Y_n[:8]
```

```python
Y_n = np.array(Y_n)
Y_n = Y_n.reshape(1400,1)
Y_n.shape
```

```python
train_data = np.append(X_np, Y_n, axis = 1)
```

```python
train_data
```

```python
df = pd.DataFrame(train_data)
```

```python
df.to_csv("train.csv", index=False)
```
