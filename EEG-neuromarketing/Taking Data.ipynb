```python
import os
import pandas as pd
import numpy as np
```

```python
data_path = "25-users/"
```

```python
user_arr = os.listdir(data_path)
```

```python
len(user_arr)
```

```python
label_path = "labels/"
```

```python
label_arr = os.listdir(label_path)
```

```python
len(label_arr)
```

```python
i1 = user_arr[0]
```

```python
arr_data = []
for i in user_arr:
    row_data = []
    with open(data_path+i) as f:
        for row in f:
            arr = row[:-1].split(' ')
            arr = list(map(int, (list(map(float, arr)))))
            row_data.append(arr)
    arr_data.append(row_data)
```

```python
arr_data = np.array(arr_data)
arr_data.shape
```

```python
arr_data = arr_data.reshape(1045, -1)
```

```python
arr_data.shape
```

```python
arr_data[0:10]
```

```python
labels = []
for j in label_arr:
    with open(label_path+j) as f:
        for r in f:
            if r == 'Disike':
                labels.append(0)
            else:
                labels.append(1)
```

```python
labels
```

```python
len(labels)
```

```python
labels = np.array(labels)
```

```python
labels = labels.reshape(-1,1)
```

```python
total_data = np.append(arr_data, labels, axis=1)
```

```python
df = pd.DataFrame(total_data)
```
